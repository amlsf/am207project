```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('white')
```

```python
# for easy access to seaborn colors
colorPalette = sns.color_palette()
snsBlue = colorPalette[0]
snsGreen = colorPalette[1]
snsRed = colorPalette[2]
snsPurple = colorPalette[3]
snsYellow = colorPalette[4]
snsAzure = colorPalette[5]
```

```python
sns.set_context('notebook', font_scale=1.25)
```

```python
xx = np.arange(4.0, 11.0, 1.0)
lxx = len(xx)
```

*Comparison - number of guesses*

```python
num_guesses_mean_knuth = np.array([4.0, 4.0, 4.0, 4.0, 6.0, 6.0, 6.0])
num_guesses_mean_rs    = np.array([3.7, 4.5, 4.8, 5.0, 5.4, 5.9, 6.6])
num_guesses_mean_ea    = np.array([3.8, 4.0, 4.8, 4.8, 5.1, 5.8, 5.8])
num_guesses_mean_e     = np.array([3.7, 4.0, 4.4, 4.8, 5.4, 5.7, 6.0])
num_guesses_mean_sab   = np.array([7.5, 8.0, 8.0, 9.2, 9.1, 8.8, 9.4])
num_guesses_mean_sae   = np.array([3.5, 4.7, 5.0, 5.3, 5.5, 6.2, 6.0])
num_guesses_mean_gab   = np.array([3.9, 4.0, 4.8, 5.0, 5.3, 6.0, 6.5])
num_guesses_mean_gae   = np.array([3.8, 4.5, 5.0, 5.5, 5.5, 6.2, 5.9])
```

```python
jitter = 0.2

plt.plot(xx + np.random.uniform(-jitter, jitter), num_guesses_mean_knuth, 'o', color=snsBlue, label="Knuth's algorithm")
plt.plot(xx + np.random.uniform(-jitter, jitter), num_guesses_mean_rs,    'o', color=snsRed, label='Random search')
plt.plot(xx + np.random.uniform(-jitter, jitter), num_guesses_mean_sab,   'o', color=snsGreen, label='Simulated annealing')
plt.plot(xx + np.random.uniform(-jitter, jitter), num_guesses_mean_gab,   'o', color=snsYellow, label='Genetic algorithms')
plt.ylim(ymin=0, ymax=10)
plt.xlim(xmin=(min(xx)-0.5), xmax=(max(xx)+0.5))
plt.xlabel('Number of possible colors (4 positions)')
plt.ylabel('Number of guesses')
plt.legend(loc='lower right')
plt.title('Optimization Strategies')
plt.show()

plt.plot(xx + np.random.uniform(-jitter, jitter), num_guesses_mean_ea,    'o', color=snsBlue, label='Exhaustive search, all steps')
plt.plot(xx + np.random.uniform(-jitter, jitter), num_guesses_mean_e,     'o', color=snsRed, label='Exhaustive search, except first step')
plt.plot(xx + np.random.uniform(-jitter, jitter), num_guesses_mean_sae,   'o', color=snsGreen, label='Simulated annealing')
plt.plot(xx + np.random.uniform(-jitter, jitter), num_guesses_mean_gae,   'o', color=snsYellow, label='Genetic algorithms')
plt.ylim(ymin=0, ymax=7)
plt.xlim(xmin=(min(xx)-0.5), xmax=(max(xx)+0.5))
plt.xlabel('Number of possible colors (4 positions)')
plt.ylabel('Number of guesses')
plt.legend(loc='lower right')
plt.title('Maximizing Shannon Entropy')
plt.show()
```
